```python
import pandas as pd
import io

df = pd.read_csv(io.BytesIO(uploaded['wbc.csv'])) # load dataset
df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df.info()
```

1.	Pisahkan antara variabel yang dapat digunakan dan variabel yang tidak dapat digunakan.

```python
df = df.drop(columns=['id'])
df = df.drop(columns=['Unnamed: 32'])
df.info()
```

2.	Lakukan proses encoding pada kolom "diagnosis".

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['diagnosis'] = le.fit_transform(df['diagnosis']) # proses encoding
df.head()
```

3.	Lakukan proses standarisasi pada semua kolom yang memiliki nilai numerik.

```python
std = StandardScaler()
for i in df.columns:
    if i != 'diagnosis':
        df[i] = std.fit_transform(df[[i]])
df.head()
```

4.	Lakukan proses stratified split data untuk membuat data latih dan data uji dengan rasio 80:20.

```python
# Split data
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.2, random_state=0, stratify=df['diagnosis'])

print(f'Jumlah label data asli:\n{df.diagnosis.value_counts()}\n')
print(f'Jumlah label data train:\n{df_train.diagnosis.value_counts()}\n')
print(f'Jumlah label data test:\n{df_test.diagnosis.value_counts()}\n')
```
